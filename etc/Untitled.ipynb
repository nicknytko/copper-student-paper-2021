```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 25
x = np.linspace(-1, 1, N)
y = np.linspace(-1, 1, N)
xx, yy = np.meshgrid(x,y)
```

```python
w = np.array([
    2*yy * (1 - (xx**2)),
    2*xx * (1 - (yy**2))
])
```

```python
plt.figure(figsize=(7,7))
plt.quiver(xx,yy,w[0],w[1])
```
